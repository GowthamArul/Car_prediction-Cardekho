```python
import pandas as pd
```

```python
df = pd.read_csv('car data.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
ls = ['Seller_Type', 'Transmission','Owner']
for i in ls:
    print(df[i].unique())
```

```python
df.isnull().sum()
```

```python
final_df = df.drop(columns=['Car_Name'])
```

```python
final_df
```

```python
final_df['Current_Year'] = [2022 for _ in range(len(df))]
```

```python
final_df
```

```python
final_df['Year_diff'] = final_df['Current_Year'] - final_df['Year']
```

```python
final_df
```

```python
final_df.drop(columns=['Year','Current_Year'], axis=1, inplace=True)
```

```python
final_df.head(5)
```

```python
final_df=pd.get_dummies(final_df, drop_first=True)
```

```python
final_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cormat = final_df.corr()
top_corr_features = cormat.index
plt.figure(figsize=(20,20))
g = sns.heatmap(final_df[top_corr_features].corr(), annot=True,cmap='RdYlGn')
```

```python
X= final_df.iloc[:,1:] # Ignoring the first index column and taking the remaining columns for the X 
y=final_df.iloc[:,0] # Taking the first index column for the y
```

```python
final_df.head(5)
```

```python
X.head()
```

```python
y.head()
```

## Feature Importance

```python
from sklearn.ensemble import ExtraTreesRegressor
model= ExtraTreesRegressor()
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

## Plot graph of feature importance for better visualization

```python
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(5).plot(kind='barh')
plt.show()
```
